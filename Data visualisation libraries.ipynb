```python
df=sns.load_dataset("tips")
```

```python
df
```

```python
df.head()
```

```python
df.info()
```

```python
df.plot()
```

```python
sns.relplot(data=df,x="total_bill",y="tip",hue="day",col="smoker",col_wrap=2) 
```

## displot

```python
sns.distplot(df.total_bill,kde=True)
```

```python
sns.barplot(data=df,x="sex",y="total_bill",hue="day")
```

## countplot

```python
sns.countplot(data=df, x="sex")
```

## joint plot

```python
sns.jointplot(data=df,x="total_bill",y="tip")
```

## box plot

```python
sns.boxplot(data=df,x="day",y="total_bill")
```

```python
#correlation
df
```

```python
corr=df.corr()
```

```python
sns.heatmap(corr,annot=True,cmap="YlGnBu")
```

## pair plot

```python
sns.pairplot(data=df,hue="sex",kind="scatter",palette="husl")
```
